```python
%matplotlib inline
%load_ext autoreload

%autoreload 2

import folium
from pcap_parser import PacketAnalyser
from scapy.all import rdpcap
import json
```

```python
FILENAME = "captures/phone_overnight.pcap"
```

```python
analyser = PacketAnalyser(source_ip="192.168.2.2")
```

```python
%%time 

for packet in rdpcap(FILENAME):
    analyser.analyse_packet(packet)
```

```python
def render_locations_on_map(analyser):
    m = folium.Map()
    for result in sorted(list(analyser.results)):
        lat, lng = result.coordinates
        folium.Marker([lat, lng], tooltip=result.city, popup=json.dumps(result._asdict())).add_to(m)
    return m

render_locations_on_map(analyser)
```
